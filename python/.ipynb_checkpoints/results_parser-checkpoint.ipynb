```python
import json
import os
import pandas as pd
import numpy as np
import math
```

```python
def sum_timers(timers): 
    sums = {}
    if type(timers) is list: 
        for i in timers:
            if i['type'] not in sums:
                sums[i['type']] = i['time']
            else:
                sums[i['type']] = sums[i['type']] + i['time']
            # TODO: Convert from ms to s?
    return sums
```

```python
def parse_result(df): 
    result = pd.DataFrame()
    for _, row in df.iterrows():
        survey_question = row['question']

        if 'preferences' in survey_question: #Preferences
            
            continue
        elif 'comments' in survey_question: # Comments
            
            continue
        elif 'confidence' not in survey_question: # Question Answer   
            question = row['question'][0]
            answer = row['question'][1]
            time = row['time']
            interactions = row['interactions']
            timers = row['timers']
            sums = sum_timers(row['timers'])
            continue
        else: # Confidence
            question = row['question'][0]
            confidence = row['question'][1]
            continue

    return result
```

```python
# Go through results dir and load all files in data frame
def read_files(): 
    directory = r'results'

    survey_results = pd.DataFrame()
    survey_count = 0

    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            with open(directory + '/' + filename, encoding='utf-8', mode='r') as current_file:
                
                text = os.path.splitext(filename)[0]
                # parse id and group from file name
                result_id = text.split('_')[1]
                group = text.split('_')[2]
                
                df = pd.read_json(current_file)
                
                print(text)
                data = parse_result(df) # parse single survey result

                survey_results = survey_results.append(df, ignore_index = True) # append to results
                survey_count+=1
        else:
            print('File not JSON', filename)
            continue
```

```python
def save_csv(): 
    survey_results.to_csv('results.csv')
```
