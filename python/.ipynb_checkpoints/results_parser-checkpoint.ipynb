```python
import json
import os
import pandas as pd
import numpy as np
```

```python
def parse_result(df): 
    for index, row in df.iterrows():
        
        question = row['question']
        time = row['time']
        interactions = row['interactions']
        timers = row['timers']
        if index == 0: # Question Answer
            print(question)
            print(time)
            print(interactions)
            print(timers)
            print('----------')
        if index == 1: # Confidence
            print(question)
            print('++++++++++')
    return ''
```

```python
# Go through results dir and load all files in data frame
def read_files(): 
    directory = r'results'

    survey_results = pd.DataFrame()
    survey_count = 0

    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            with open(directory + '/' + filename, encoding='utf-8', mode='r') as current_file:
                df = pd.read_json(current_file)
                data = parse_result(df) # parse single survey result
#                 data['ID'] = os.path.splitext(filename)[0].replace('survey_results_', '') # add id to results
#                 survey_results = survey_results.append(df, ignore_index = True) # append to results
#                 survey_count+=1
        else:
            print('File not JSON', filename)
            continue
```

```python
def save_csv(): 
    survey_results.to_csv('results.csv')
```
